# Concentric Data

Notebook to generate circularly classed data used to demonstrate power of kernel based PCA.

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate feature and target sets with specified parameters
X, y = make_circles(n_samples=500, factor=.2, noise=.2)

# Combine data to one object
all_data = np.c_[X, y]
```

```python
# Center the data for visualisation (unclear in docs how it is bounded)
X = X - np.mean(X, axis=0)
```

## Agglomerative Clustering Simplification

```python
# Generate feature and target sets with specified parameters
X, y = make_circles(n_samples=300, factor=.31, noise=0.07, random_state=123)

# Combine data to one object
all_data = np.c_[X, y]

X = X - np.mean(X, axis=0)
```

```python
plt.scatter(X[:,0], X[:,1], c=y);
```

```python
np.savetxt("../data/circles_simplified.csv", all_data, delimiter=",")
```
