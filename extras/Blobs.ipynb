# Blobs data generation

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
centers = [[1, 1], [-1, -1], [1, -1]]
X, y = make_blobs(n_samples=300, n_features=2, centers=centers, cluster_std=0.4)
X.shape
```

```python
plt.scatter(X[:,0], X[:,1], c=y);
```

```python
data = np.c_[X, y]
#np.savetxt("../data/clusters.csv", data, delimiter=",")
```

## Create a new different density set of blobs

```python
centers = [[1, 1], [-1, -1], [1, -1]]
X, y = make_blobs(n_samples=500, n_features=2, centers=centers, cluster_std=0.6)
X.shape
```

```python
plt.scatter(X[:,0], X[:,1], c=y);
```

```python
data = np.c_[X, y]
data0 = data[data[:,-1] == 0]
data1 = data[data[:,-1] == 1]
data2 = data[data[:,-1] == 2]
```

```python
proportion = 0.7

data2 = data2[np.random.randint(low=0, high=data2.shape[0], size=int(data2.shape[0]*proportion))]
```

```python
data2.shape
```

```python
data = np.concatenate([data0, data1, data2])
data.shape
```

```python
plt.scatter(data[:,0], data[:,1], c=data[:,-1]);
```

```python
#np.savetxt("../data/clusters_density.csv", data, delimiter=",")
```

# Exercise 1 & 3 data production

```python
centers = [[1.1, 0.9], [0.2, 0], [-1.1, -1], [1.05, -1], [-1.019, 0.8]]
X, y = make_blobs(n_samples=[43, 95, 56, 112, 74], 
                  n_features=2, 
                  centers=centers, 
                  cluster_std=[0.3, 0.25, 0.4, 0.35, 0.3], 
                  random_state=1)
X.shape
```

```python
plt.scatter(x=X[:,0], y=X[:,1])
```

```python
data = np.c_[X, y]
np.savetxt("../data/exercise_one_clusters.csv", data, delimiter=",")
np.savetxt("../data/exercise_three_clusters.csv", data, delimiter=",")
```
