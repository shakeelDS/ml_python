```python
from sklearn.datasets import make_swiss_roll

n_samples = 2000
noise = 0.5
X, y = make_swiss_roll(n_samples, noise, random_state=1)

# Move centrally
X[:, 1] = (X[:, 1] - np.mean(X[:, 1])) / np.std(X[:, 1])
X[:, 0] = (X[:, 0] - np.mean(X[:, 0])) / np.std(X[:, 0])
X[:, 2] = (X[:, 2] - np.mean(X[:, 2])) / np.std(X[:, 2])

plt.scatter(x=X[:,0], y=X[:,1])
```

```python
# Remove middle
X_1 = X[(X[:, 1] > 1)]
X_2 = -X[(X[:,1] < -1)]
```

```python
# Shift 
X_1[:,1] = X_1[:,1] - 1.2
X_1[:,0] = X_1[:,0] + 0.5
```

```python
plt.scatter(x=X_1[:,0], y=X_1[:,1])
plt.scatter(x=X_2[:,0], y=X_2[:,1])
```

```python
plt.scatter(x=X_1[:,0], y=X_1[:,2])
plt.scatter(x=X_2[:,0], y=X_2[:,2])
```

```python
y_1 = np.full(
    shape=X_1.shape[0],
    fill_value=0,
    dtype=np.int)

y_2 = np.full(
    shape=X_2.shape[0],
    fill_value=1,
    dtype=np.int)
```

```python
X = np.concatenate((X_1, X_2))

# re-order the data columns
X = X[:,[0, 2, 1]]
y = np.concatenate((y_1, y_2))

all_data = np.c_[X, y]

```
